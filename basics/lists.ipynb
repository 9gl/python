### Patterns of looping

```python
for student in classroom:
  print(student)
```

```python
for student in reversed(classroom):
  print(student)
```

```python
for rollno, student in enumerate(classroom):
  print(rollno, student)
```

```python
for rollno, student in enumerate(classroom, start = 1):
  print(rollno, student)
```

```python
def square_list(numbers):
  sqrs = []
  for n in numbers:
    sqrs.append(n*n)
  return sqrs
```

### Problems

- Write a list comprehension to generate multiplication tables of 1-5
- Write a function to find all factors of given number ,make use of list comprehension
- Find sum of all integers divisible by 13 or 17 and less than 1000 sum() -> can be used

```python
# Q1.
numbers = list(range(1,5))
one_ten = list(range(1,10))
def table(n)->int:
  for x in range(1,10):
    return x*n
for x in numbers:
  y = table(x)
  print(y)
```

```python
# Q2.
def factors(n):
  return [i for i in range(1,n+1) if n%i == 0]
x = factors(10)
print(x)
```

```python
# Q3.
def req_question():
  lst  = [i for i in range(1,1000) if i%13 == 0 or i%17 ==0]
  return sum(lst)
```

###Bonus problem
- From 2D list, extract nth column
...
column(list2d, 2) -> 2nd column  

```python
def column(list2d, n) -> list:
    index = n - 1
    return [list2d[x][index] for x in range(len(list2d))]
```
