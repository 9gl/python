### Problem
-make use of above concept to sum pi series 
8/(1*3) + 8/(5*7) + 8/(7*9) ... = pi

```python
def sumterms(start, end, step,term):
    s = 0
    for n in range(start, end, step):
      s += term(n)
    return s
```

```python
def piseries(n):
  return 8/((4*n-2)*(4*n-1))
```

```python
sumterms(1, 1000,1, piseries)
```
