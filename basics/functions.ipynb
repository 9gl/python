### Problem

8/(1*3) + 8/(5*7) + 8/(7*9) ... = pi

```python
def sumterms(start, end, step,term):
    s = 0
    for n in range(start, end, step):
      s += term(n)
    return s
```

```python
def piseries(n):
  return 8/((4*n-2)*(4*n-1))
```

```python
sumterms(1, 1000,1, piseries)
```

```python
def column(data2d, col):
    return [data2d[r][col] for r in range(len(data2d))]
```

```python
def transpose(data2d):
    colcount = len(data2d[0])
    return [column(data2d, c) for c in range(colcount)]
```
